```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Datos extrados de la tabla
absolute_pre = [
    0.21, 0.41, 0.31, 0.27, 0.39, 0.3, 0.17, 0.09, 0.19, 0.68,
    0.21, 0.43, 0.29, 0.99, 0.19, 0.37, 0.3, 0.13, 0.75, 0.21,
    0.37, 0.55, 1.03, 1.76
]

relative_pre = [
    0.02, 0.07, 0.1, 0.05, 0.04, 0.09, 0.08, 0.02, 0.03, 0.09,
    0.06, 0.17, 0.03, 0.26, 0.02, 0.09, 0.1, 0.02, 0.14, 0.12,
    0.09, 0.14, 0.09, 0.24
]

absolute_post = [
    0.44, 0.37, 0.22, 0.26, 0.73, 1.0, 0.26, 0.24, 0.44, 1.38,
    0.31, 0.56, 0.38, 0.46, 0.19, 0.31, 0.26, 0.28, 0.32, 0.73,
    0.45, 0.83, 0.28, 0.39
]

relative_post = [
    0.04, 0.05, 0.11, 0.11, 0.09, 0.1, 0.05, 0.03, 0.04, 0.38,
    0.08, 0.25, 0.04, 0.06, 0.02, 0.05, 0.03, 0.05, 0.05, 0.1,
    0.09, 0.1, 0.05, 0.04
]

# Crear los histogramas
plt.figure(figsize=(15, 10))

# Histograma para Absolute - pre
plt.subplot(2, 2, 1) # 2 filas, 2 columnas, primer grfico
plt.hist(absolute_pre, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribucin de "Absolute - pre (mm)"')
plt.xlabel('Valor (mm)')
plt.ylabel('Frecuencia')

# Histograma para Relative - pre
plt.subplot(2, 2, 2) # Segundo grfico
plt.hist(relative_pre, bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribucin de "Relative - pre (mm)"')
plt.xlabel('Valor (mm)')
plt.ylabel('Frecuencia')

# Histograma para Absolute - post
plt.subplot(2, 2, 3) # Tercer grfico
plt.hist(absolute_post, bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribucin de "Absolute - post (mm)"')
plt.xlabel('Valor (mm)')
plt.ylabel('Frecuencia')

# Histograma para Relative - post
plt.subplot(2, 2, 4) # Cuarto grfico
plt.hist(relative_post, bins=10, color='gold', edgecolor='black')
plt.title('Distribucin de "Relative - post (mm)"')
plt.xlabel('Valor (mm)')
plt.ylabel('Frecuencia')

plt.tight_layout() # Ajustar el diseo para evitar superposiciones
plt.show()
```

```python

# Crear los grficos de dispersin
plt.figure(figsize=(14, 6))

# Grfico de dispersin para Pre
plt.subplot(1, 2, 1) # 1 fila, 2 columnas, primer grfico
plt.scatter(absolute_pre, relative_pre, color='blue', alpha=0.7)
for i in range(len(absolute_pre)):
    plt.annotate(f'S {i+1}', (absolute_pre[i], relative_pre[i]), textcoords="offset points", xytext=(0,5), ha='center', fontsize=8)
plt.title('Valores "Pre" por Sujeto (Absolute vs Relative)')
plt.xlabel('Absolute - pre (mm)')
plt.ylabel('Relative - pre (mm)')
plt.grid(True, linestyle='--', alpha=0.6)

# Grfico de dispersin para Post
plt.subplot(1, 2, 2) # Segundo grfico
plt.scatter(absolute_post, relative_post, color='green', alpha=0.7)
for i in range(len(absolute_post)):
    plt.annotate(f'S {i+1}', (absolute_post[i], relative_post[i]), textcoords="offset points", xytext=(0,5), ha='center', fontsize=8)
plt.title('Valores "Post" por Sujeto (Absolute vs Relative)')
plt.xlabel('Absolute - post (mm)')
plt.ylabel('Relative - post (mm)')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python

# Nmero de sujetos
num_sujetos = len(absolute_pre)
sujetos = np.arange(num_sujetos) # Eje X para las barras

bar_width = 0.2 # Ancho de cada barra

plt.figure(figsize=(18, 8))

# Posiciones de las barras para cada grupo
r1 = sujetos - 1.5 * bar_width
r2 = sujetos - 0.5 * bar_width
r3 = sujetos + 0.5 * bar_width
r4 = sujetos + 1.5 * bar_width

plt.bar(r1, absolute_pre, color='skyblue', width=bar_width, edgecolor='grey', label='Absolute - Pre')
plt.bar(r2, relative_pre, color='lightcoral', width=bar_width, edgecolor='grey', label='Relative - Pre')
plt.bar(r3, absolute_post, color='lightgreen', width=bar_width, edgecolor='grey', label='Absolute - Post')
plt.bar(r4, relative_post, color='gold', width=bar_width, edgecolor='grey', label='Relative - Post')

plt.xlabel('Sujeto', fontweight='bold')
plt.ylabel('Valor (mm)', fontweight='bold')
plt.title('Valores "Absolute" y "Relative" por Sujeto (Pre y Post)', fontweight='bold')
plt.xticks(sujetos, [f'Sujeto {i+1}' for i in range(num_sujetos)], rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
